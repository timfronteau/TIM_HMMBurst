```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
target_dir = "tde-hmm2/tcourses-comparison/"
```

```python
subj_list = [i for i in range(2,9)] + [i for i in range(10,13)] + [14, 23]
```

```python
n_trials = 5
```

### Script

```python
for n_states in range(3,7):
    for subj in subj_list:
        fig, f_axes = plt.subplots(figsize=(20,n_states*n_trials), ncols=1, nrows=(n_states*n_trials), constrained_layout=True)
        ds = xr.open_dataset(f"tde-hmm2/nc_files/su{subj}-{n_states}states_data_Multi.nc")
        for state in range(1, n_states+1):
            color = f"C{state-1}"
            for trial in range(n_trials):
                ax = f_axes[n_trials*(state-1)+trial]
                ax.fill_between(ds["time"], ds["states_timecourse"].values[trial, :, state-1], color=color)
                ax.set_xlim([-4, 3])
                ax.set_ylim([0, 1])
                ax.set_ylabel(f"Prob. st{state} tr{trial}")
        ax.set_xlabel("Time (s)")
        fig.savefig(target_dir + f'tcourses-{n_states}states-subj{subj}.png', dpi=300)
        plt.close(fig)
```
