```python
import mne
from wavelet_transform import wavelet_transform
```

```python
csvfile = open('FCK_LOCKED_IC_TIM_tfr256Hz_11062021.csv', 'a', newline='')
```

```python
spamwriter = csv.writer(csvfile, delimiter=';',
                        quotechar='|', quoting=csv.QUOTE_MINIMAL)
```

```python
spamwriter
```

```python
info = mne.create_info(ch_names=['signal'], sfreq=256, ch_types=['eeg'])
```

```python
type(info)
```

```python
freq_shape = data['freq_axis'].shape[0]
print(freq_shape)
time_shape = data['time_axis'].shape[0]
print(time_shape)
```

```python
print("Computing and loading the time-frequency wavelet transformation of three subjects for all IC, all trials...")
for subj in range(2,5):
    for trial in range(1, n_trials+1):
        for IC in range(1, n_IC+1):
            try:
                wtfr = wavelet_transform(data, info, subj, IC, trial)
                for i in range(freq_shape):
                    spamwriter.writerow([f'tfr_256Hz subject{subj}, IC{IC}, trial{trial}, f_ind{i}'] 
                                        + [ wtfr[0, i, j] for j in range(time_shape) ])
            except:
                pass
        print(f'Trial{trial} loaded')
print("Loaded")
```
